```python
from numpy import arange, array, cos, log, pi, sin, sqrt

from math import pi
from bokeh.models import Title
from bokeh.palettes import Category20c
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, output_file, show
from bokeh.transform import cumsum
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource, HoverTool, LabelSet
import itertools
```

```python
# Ensure the DataFrame is properly structured
df['angle'] = df['LowestAccess_HalfM'] / df['LowestAccess_HalfM'].sum() * 2 * np.pi

# Define a color palette (modify or extend if needed)
color_palette = ['#B0C4DE', '#BC8F8F', '#CD5C5C', '#DEB887', '#9467bd', '#8c564b', '#e377c2']

# Get unique race categories
unique_races = df['RaceWithLowestAccess'].unique()

# Create a color mapping (each race gets a unique color)
race_color_map = {race: color for race, color in zip(unique_races, itertools.cycle(color_palette))}

# Assign colors to the DataFrame based on the race category
df['color'] = df['RaceWithLowestAccess'].map(race_color_map)

# Map the race categories to full labels for the legend
race_label_map = {
    'NHPI': 'Native Hawaiian or Other Pacific Islander',
    'NAAN': 'American Indian or Alaska Native'
}
df['RaceWithLowestAccessLabel'] = df['RaceWithLowestAccess'].map(race_label_map).fillna(df['RaceWithLowestAccess'])

# ðŸ”¹ Add scaled outer_radius column
df['outer_radius'] = 0.2 + df['LowestAccess_HalfM'] / df['LowestAccess_HalfM'].max() * 0.4

# ðŸ”¹ Add a formatted column for hover display
df['LowestAccess_HalfM_str'] = df['LowestAccess_HalfM'].astype(str)

# ðŸ”¹ Calculate the percentage for each race
df['Percentage'] = df['LowestAccess_HalfM'] / df['LowestAccess_HalfM'].sum() * 100
df['Percentage_str'] = df['Percentage'].round(2).astype(str) + "%"

source = ColumnDataSource(df)

# Create the figure
p = figure(
    title="Low Food Access by Race (Within 1/2 Mile)",
    x_range=(-1, 1), y_range=(-1, 1),
    height=600, width=600,
    tools="pan,wheel_zoom,reset"  # Enable interactive tools
)

# Plot the annular wedges
p.annular_wedge(
    x=0, y=0,
    inner_radius=0.2,
    outer_radius='outer_radius',  # Reference from source
    start_angle=cumsum('angle', include_zero=True),
    end_angle=cumsum('angle'),
    line_color="white",
    fill_color='color',
    legend_field="RaceWithLowestAccessLabel",  # Use updated label column for legend
    source=source
)

# Add a hover tool to display information on hover
hover = HoverTool()
hover.tooltips = [
    ("Race", "@RaceWithLowestAccessLabel"),
    ("Lowest Access", "@LowestAccess_HalfM_str"),
    ("Percentage", "@Percentage_str"),
    ("State", "@State"),  # Show State
    ("County", "@County")  # Show County
]
p.add_tools(hover)

# Hide axis lines for a cleaner look
p.axis.visible = False

# Save the interactive plot as an HTML file
output_file("interactive_food_access_by_race.html")

# Display the plot
show(p)
```

```python
### Breakdown of the SQL Query:
SELECT Clause:
The query selects the State and County columns from the food_deserts table.
It calculates the proportion of each racial group's population with low food access. This is done by dividing the sum of people from each racial group with low access to food by the total population of that racial group:
WhiteLowAccessShare
BlackLowAccessShare
AsianLowAccessShare
NHPILowAccessShare (Native Hawaiian and Pacific Islander)
NAANLowAccessShare (Native American and Alaskan Native)
OtherLowAccessShare
HispLatamLowAccessShare (Hispanic and Latin American)
This is done for each racial group by using the SUM() function, which adds up the number of people with low access for that race. Each sum is cast to a FLOAT to ensure decimal precision, and the result is divided by the total population for that group.
CASE Statement:
The query uses a CASE statement to determine which racial group has the highest proportion of people with low food access for each county/state.
For each race, it compares the calculated LowAccessShare for each group (like White, Black, Asian, etc.) and determines which race has the largest share. The race with the highest proportion is returned as the RaceWithLowestAccess column.
For example:
If the WhiteLowAccessShare is the highest, the query will return 'White' in the RaceWithLowestAccess column for that particular state/county.
This comparison continues for all the racial groups listed in the query.
FROM Clause:
The query pulls data from the food_deserts table.
GROUP BY Clause:
The query groups the data by State and County, so the calculations and the CASE statement are done per state/county.
ORDER BY Clause:
The result is ordered by WhiteLowAccessShare in descending order (DESC). This means the states/counties where White populations have the highest share of low access to food will appear first.
LIMIT Clause:
The query limits the result to only one row. Therefore, the result will show the state and county with the highest WhiteLowAccessShare.
```
